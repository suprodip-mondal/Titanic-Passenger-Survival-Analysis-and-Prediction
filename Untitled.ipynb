## Data Cleaning

```python
plt.figure(figsize=(12, 7))
sns.boxplot(x='Sex',y='Age',hue='Pclass',data=df,palette='winter')
```

```python
df.groupby(['Sex', 'Pclass'])['Age'].mean()
```

```python
def age(cols):
    Age = cols[0]
    Pclass = cols[1]
    Sex = cols[2]
    
    if pd.isnull(Age):
        if Pclass == 1:
            return 35 if Sex == 'Female' else 41
        elif Pclass == 2:
            return 29 if Sex == 'Female' else 31
        else: # Pclass == 3
            return 22 if Sex == 'Female' else 27
    else:
        return Age
```

```python
df['Age'] = df[['Age','Pclass','Sex']].apply(age,axis=1)
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df.drop('Cabin',axis=1,inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

## Converting into Categorical features

```python
sex = pd.get_dummies(df['Sex'],drop_first=True)
embark = pd.get_dummies(df['Embarked'],drop_first=True)
```

```python
df.drop(['Sex','Embarked','Name','Ticket'],axis=1,inplace=True)
```

```python
df = pd.concat([df,sex,embark],axis=1)
```

```python
df.head()
```

## Bulging a Logistic Regression Model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.metrics import classification_report,confusion_matrix,recall_score,f1_score,precision_score,roc_curve, roc_auc_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop('Survived',axis=1), df['Survived'], test_size=0.30)
```

```python
lr = LogisticRegression()
rf = RandomForestClassifier()
gb = GradientBoostingClassifier()
```

```python
models = [lr,rf,gb]

for i in models:
    i.fit(x_train,y_train)
```

```python
lr_pred = lr.predict(x_test)
rf_pred = rf.predict(x_test)
gb_pred = gb.predict(x_test)
```

```python
lr_precision = precision_score(y_test, lr_pred)
lr_recall = recall_score(y_test, lr_pred)
lr_f1 = f1_score(y_test, lr_pred)

rf_precision = precision_score(y_test, rf_pred)
rf_recall = recall_score(y_test, rf_pred)
rf_f1 = f1_score(y_test, rf_pred)

gb_precision = precision_score(y_test, gb_pred)
gb_recall = recall_score(y_test, gb_pred)
gb_f1 = f1_score(y_test, gb_pred)
```

```python
comparison_data = {
    "Metric": ["Precision", "Recall", "F1-Score"],
    "Logistic Regression": [lr_precision, lr_recall, lr_f1],
    "Random Forest": [rf_precision, rf_recall, rf_f1],
    "Gradient Boosting": [gb_precision, gb_recall, gb_f1]
}
```

```python
pd.DataFrame(comparison_data)
```

```python
predictions = [lr_pred,rf_pred,gb_pred]
model_names = ["Logistic Regression", "Random Forest", "Gradient Boosting"]

# 3. Use zip to loop through both lists at the same time
for name, pred in zip(model_names, predictions):
    print(f"--- Classification Report for: {name} ---")
    print(classification_report(y_test, pred))
    print("="*60 + "\n")
```

```python
lr_y_pred_proba = lr.predict_proba(x_test)[:, 1]
rf_y_pred_proba = rf.predict_proba(x_test)[:, 1]
gb_y_pred_proba = gb.predict_proba(x_test)[:, 1]
```

```python
lr_fpr, lr_tpr, lr_thresholds = roc_curve(y_test, lr_y_pred_proba)
lr_auc_score = roc_auc_score(y_test, lr_y_pred_proba)

rf_fpr, rf_tpr, rf_thresholds = roc_curve(y_test, rf_y_pred_proba)
rf_auc_score = roc_auc_score(y_test, rf_y_pred_proba)

gb_fpr, gb_tpr, gb_thresholds = roc_curve(y_test, gb_y_pred_proba)
gb_auc_score = roc_auc_score(y_test, gb_y_pred_proba)
```

```python
plt.figure(figsize=(12,8))
plt.plot(lr_fpr, lr_tpr, color='lightgreen', lw=2, label=f'Logistic Regression (AUC = {lr_auc_score:.2f})')

plt.plot(rf_fpr, rf_tpr, color='darkorange', lw=2, label=f'Random Forest (AUC = {rf_auc_score:.2f})')

plt.plot(gb_fpr, gb_tpr, color='navy', lw=2, label=f'Gradient Boosting (AUC = {gb_auc_score:.2f})')

plt.plot([0, 1], [0, 1], color='red', lw=2, linestyle='--')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.title('ROC Curve Comparison: Random Forest vs. Logistic Regression Vs. Gradient Boosting')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```
